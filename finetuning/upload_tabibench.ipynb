```python
# 3. PosUDBOUN
pos_boun_train_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_boun/tr_boun-ud-train.conllu"
pos_boun_test_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_boun/tr_boun-ud-test.conllu"
pos_boun_dev_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_boun/tr_boun-ud-dev.conllu"

from conllu import parse_incr
def load_conllu_to_dataset(filepath):
    sentences = []
    with open(filepath, "r", encoding="utf-8") as f:
        for tokenlist in parse_incr(f):
            tokens = [token["form"] for token in tokenlist]
            lemmas = [token["lemma"] for token in tokenlist]
            pos_tags = [token["upostag"] for token in tokenlist]

            sentences.append({
                "tokens": tokens,
                "lemmas": lemmas,
                "pos_tags": pos_tags
            })
    return Dataset.from_list(sentences)

train_dataset = load_conllu_to_dataset(pos_boun_train_path)
eval_dataset = load_conllu_to_dataset(pos_boun_dev_path)
test_dataset = load_conllu_to_dataset(pos_boun_test_path)

boun_ds = DatasetDict({
    'train': train_dataset,
    'validation': eval_dataset,
    'test': test_dataset,
})
print(boun_ds)
boun_ds.push_to_hub("boun-tabilab/PosUD-BOUN")

# 4. PosUDIMST
pos_imst_train_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_imst/tr_imst-ud-train.conllu"
pos_imst_test_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_imst/tr_imst-ud-test.conllu"
pos_imst_dev_path = "/workspace/TabiBERT/finetuning/datasets/pos_ud_imst/tr_imst-ud-dev.conllu"

train_dataset = load_conllu_to_dataset(pos_imst_train_path)
eval_dataset = load_conllu_to_dataset(pos_imst_dev_path)
test_dataset = load_conllu_to_dataset(pos_imst_test_path)

imst_ds = DatasetDict({
    'train': train_dataset,
    'validation': eval_dataset,
    'test': test_dataset,
})
print(imst_ds)
imst_ds.push_to_hub("boun-tabilab/PosUD-IMST")
```

# STS							

```python
from dataloaders.stsb_tr import SICKReader, STSBReader

# 1. SICK-TR
path_sick = "/workspace/TabiBERT/finetuning/datasets/sick_tr/"
sick_reader = SICKReader(path_sick)
sick_dataset = sick_reader.get_datasets()

train_dataset = sick_dataset.get("train")
validation_dataset = sick_dataset.get("trial")
test_dataset = sick_dataset.get("test")

columns_to_remove = ["id", "entailment_label", "label"]

def safe_remove_columns(ds, cols):
    # Only remove columns that exist in the dataset
    existing_cols = [c for c in cols if c in ds.column_names]
    return ds.remove_columns(existing_cols)

sick_ds = DatasetDict({
    "train": safe_remove_columns(train_dataset, columns_to_remove),
    "validation": safe_remove_columns(validation_dataset, columns_to_remove),
    "test": safe_remove_columns(test_dataset, columns_to_remove),
})
print("\nSICK-TR Dataset Loaded:")
print(sick_ds)
for split, data in sick_ds.items():
    print(f"- {split}: {len(data)} samples, columns={data.column_names}")
sick_ds.push_to_hub("boun-tabilab/SICK-TR")

# 2. STSb-TR
path_STSB = '/workspace/TabiBERT/finetuning/datasets/stsb_tr/'
stsb_reader = STSBReader(path_STSB)
sts_dataset = stsb_reader.get_datasets()

train_dataset, eval_dataset, test_dataset = sts_dataset['train'], sts_dataset[
    'dev'], sts_dataset['test']

columns_to_remove = ["id", "year"]

sts_ds = DatasetDict({
    "train": safe_remove_columns(train_dataset, columns_to_remove),
    "validation": safe_remove_columns(eval_dataset, columns_to_remove),
    "test": safe_remove_columns(test_dataset, columns_to_remove),
})
print("STSb-TR Dataset:")
print(sts_ds)
sts_ds.push_to_hub("boun-tabilab/STSb-TR")
```

# Retrieval

```python
retrieval_tasks = {
    'Wmt16-TR': 'trmteb/wmt16_en_tr_fine_tuning_dataset',
    'MsMarco-TR': 'trmteb/msmarco-tr_fine_tuning_dataset',
    'Scifact-TR': 'trmteb/scifact-tr_fine_tuning_dataset',
    'Fiqa-TR': 'trmteb/fiqa-tr_fine_tuning_dataset',
    'NFCorpus-TR': 'trmteb/nfcorpus-tr_fine_tuning_dataset',
    'Quora-TR': 'trmteb/quora-tr_fine_tuning_dataset',
}

for key, value in retrieval_tasks.items():
    print(f"Dataset in {key}")
    dataset = load_dataset(value)
    if len(dataset.keys()) == 3:
        if 'dev' in dataset:
            dataset["validation"] = dataset.pop("dev")
    elif len(dataset.keys()) == 2:
        if 'train' in dataset and 'test' in dataset:
            ratio = int(dataset['test'].num_rows) / int(dataset['train'].num_rows)
            train_dev = dataset['train'].train_test_split(test_size=ratio, seed=25)
            dataset["train"] = train_dev['train']
            dataset["validation"] = train_dev['test']
        elif 'dev' in dataset and 'test' in dataset:
            train_dev = dataset['dev'].train_test_split(test_size=0.2, seed=25)
            train = train_dev['train']
            dev = train_dev['test']
            dataset['train'] = train
            dataset['validation'] = dev
            dataset.pop("dev")
    print(dataset)
    print('\n')
    dataset.push_to_hub(f"boun-tabilab/{key}")
```

```python
# Code Retrieval:
dataset_path = [
    "boun-tabilab/Apps-TR",
    "boun-tabilab/CosQA-TR",
    "boun-tabilab/StackoverflowQA-TR",
    "boun-tabilab/CodeSearchNet-21K-TR",
]
for path in dataset_path:
    dataset = load_dataset(path)
    if 'dev' in dataset:
        dataset['validation'] = dataset.pop('dev')
    elif 'validation' not in dataset:
        ratio = int(dataset['test'].num_rows) / int(dataset['train'].num_rows)
        train_dev = dataset['train'].train_test_split(test_size=ratio, seed=25)
        dataset["train"] = train_dev['train']
        dataset["validation"] = train_dev['test']
    print(f"Dataset in {path}")
    print(dataset)
    print('\n')
    dataset.push_to_hub(path)
```
